```python
new_df = new_df.drop(columns = ['5','6','7','tot2018'])
```

```python
new_df['tot18'] = new_df['lv1']+new_df['lv2']+new_df['lv3']+new_df['lv4']+new_df['lv5']
```

```python
new_df['lv1'] = new_df['lv1']/new_df['tot18']
```

```python
new_df['lv2'] = new_df['lv2']/new_df['tot18']
new_df['lv3'] = new_df['lv3']/new_df['tot18']
new_df['lv4'] = new_df['lv4']/new_df['tot18']
new_df['lv5'] = new_df['lv5']/new_df['tot18']
```

```python
new_df
```

```python
new_df['avg_s'] =new_df['lv2']*6+new_df['lv3']*9+new_df['lv4']*12+new_df['lv5']*16
```

```python
def edu_gini(data, y):
    mu = 1/data['avg_s']
    g = 0
    for i in range(1,5):
        yi = y[i]
        p = 0
        for j in range(0,i-1):
            yj = y[j]
            p += data['lv'+str(i+1)]*abs(yi-yj)*data['lv'+str(j+1)]
        g += p
    g = g*mu
    return g
```

```python
new_df['edugini'] = edu_gini(new_df,[0,6,9,12,16])
```
